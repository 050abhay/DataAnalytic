# Subplots in Matplotlib

Subplots in Matplotlib allow you to create multiple plots within a single figure, enabling you to visualize different aspects of your data side by side. This is particularly useful when you want to compare different datasets, plot different views of the same data, or create complex layouts of plots. Subplots are versatile and can be customized in various ways, including adjusting their layout, sharing axes, and adding annotations.

## 1. Basic Usage of Subplots

The simplest way to create subplots is using the `plt.subplot()` function, which allows you to create a grid of plots within a single figure.

### a. Creating a 2x2 Grid of Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a 2x2 grid of subplots
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)  # (rows, columns, index)
plt.plot(x, y1)
plt.title('Sine Wave')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Cosine Wave')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Tangent Wave')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Exponential')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Display the plot
plt.show()
```

In this example:
- `plt.subplot(2, 2, 1)` creates the first subplot in a 2x2 grid.
- The `tight_layout()` function is used to automatically adjust subplot parameters to give specified padding.

## 2. Using `plt.subplots()` for More Control

The `plt.subplots()` function is more powerful and flexible than `plt.subplot()`. It returns a figure object and an array of axes objects, allowing you to control the layout and appearance of subplots more precisely.

### a. Creating a 2x2 Grid with plt.subplots()

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plotting data on each subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Wave')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Wave')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tangent Wave')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exponential')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

In this example:
- `fig, axes = plt.subplots(2, 2)` creates a 2x2 grid of subplots, with `axes` being an array of Axes objects.
- `axes[i, j]` refers to each subplot, allowing you to easily customize and plot data on each one.

## 3. Customizing Subplot Layouts

You can customize the layout of subplots to fit your specific needs, including creating grids of different sizes, sharing axes, and adjusting spacing.

### a. Creating Subplots with Different Sizes

You can create subplots with different sizes by specifying the grid size and the positions of each subplot.

```python
# Create a figure with subplots of different sizes
fig = plt.figure(figsize=(8, 6))

# Main plot spanning two columns
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot(x, y1)
ax1.set_title('Main Plot')

# Small subplot
ax2 = plt.subplot2grid((3, 3), (1, 0))
ax2.plot(x, y2)
ax2.set_title('Small Subplot 1')

# Small subplot
ax3 = plt.subplot2grid((3, 3), (1, 1))
ax3.plot(x, y3)
ax3.set_title('Small Subplot 2')

# Small subplot spanning two columns
ax4 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4.plot(x, y4)
ax4.set_title('Tall Subplot')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

In this example:
- `plt.subplot2grid((3, 3), (0, 0), colspan=3)` creates a subplot that spans three columns in a 3x3 grid.
- `plt.subplot2grid((3, 3), (1, 2), rowspan=2)` creates a subplot that spans two rows.

### b. Sharing Axes Between Subplots

Sharing axes between subplots ensures that the subplots use the same scale, which is particularly useful for comparison.

```python
# Create subplots with shared x-axis
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

axes[0].plot(x, y1)
axes[0].set_title('Sine Wave')

axes[1].plot(x, y2)
axes[1].set_title('Cosine Wave')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

In this example:
- `sharex=True` ensures that the x-axis is shared between the subplots.
You can also share the y-axis using `sharey=True`.

## 4. Adjusting Subplot Spacing

Matplotlib provides several methods to adjust the spacing between subplots, ensuring that they are not too close together or too far apart.


### a. Using `plt.tight_layout()`
The `tight_layout()` function automatically adjusts subplot parameters to ensure that subplots fit within the figure area without overlapping.


```python

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plotting data on each subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Wave')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Wave')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tangent Wave')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exponential')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()

```


#### b. Using plt.subplots_adjust()
The `subplots_adjust()` function gives you fine control over the spacing between subplots.


```python

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plotting data on each subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Wave')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Wave')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tangent Wave')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exponential')

# Adjust spacing between subplots
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.3, hspace=0.4)

# Display the plot
plt.show()

```


In this example:
- `wspace` and `hspace` control the amount of width and height reserved for space between subplots.
- `left`, `right`, `top`, `bottom` adjust the margins around the subplots.


## 4. Adjusting Subplot Spacing

Matplotlib provides several methods to adjust the spacing between subplots, ensuring that they are not too close together or too far apart.

### a. Adjusting Spacing with `subplots_adjust()`

```python
# Create subplots with adjusted spacing
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Wave')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Wave')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tangent Wave')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exponential')

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.5)

# Display the plot
plt.show()
```

In this example:
- `plt.subplots_adjust(wspace=0.3, hspace=0.5)` adjusts the width and height spacing between the subplots.

### b. Creating Complex Layouts with GridSpec

Matplotlib's `GridSpec` is a powerful tool for creating complex subplot layouts. It allows you to create subplots that span multiple rows or columns with more control over spacing.

```python
import matplotlib.gridspec as gridspec

# Create a GridSpec layout
fig = plt.figure(figsize=(10, 8))
gs = gridspec.GridSpec(3, 3)

# Main plot spanning two columns
ax1 = fig.add_subplot(gs[0, :2])
ax1.plot(x, y1)
ax1.set_title('Main Plot')

# Tall subplot
ax2 = fig.add_subplot(gs[1:, -1])
ax2.plot(x, y2)
ax2.set_title('Tall Subplot')

# Small subplots
ax3 = fig.add_subplot(gs[1, :-1])
ax3.plot(x, y3)
ax3.set_title('Small Subplot 1')

ax4 = fig.add_subplot(gs[2, :-1])
ax4.plot(x, y4)
ax4.set_title('Small Subplot 2')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

In this example:
- `GridSpec(3, 3)` creates a 3x3 grid layout.
- `fig.add_subplot(gs[0, :2])` creates a subplot that spans the first row and two columns.

## 5. Adding Annotations to Subplots

Annotations can be added to subplots to highlight specific data points, add labels, or create custom notes within your plots.

### a. Adding Annotations Example

```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Sine wave with annotation
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Wave')
axes[0, 0].annotate('Local Max', xy=(1.57, 1), xytext=(3, 1.5),
                 arrowprops=dict(facecolor='black', arrowstyle='->'))

# Cosine wave with annotation
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].annotate('Local Min', xy=(4.71, -1), xytext=(6, -1.5),
                 arrowprops=dict(facecolor='blue', arrowstyle='->'))

# Tangent wave without annotation
axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tangent Wave')

# Exponential with annotation
axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exponential')
axes[1, 1].annotate('Steep Rise', xy=(2, np.exp(2)), xytext=(3, 50),
                 arrowprops=dict(facecolor='red', arrowstyle='->'))

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

In this example:
- `annotate()` is used to add annotations to specific points on the plot.
- The `arrowprops` parameter customizes the appearance of the annotation arrow.


## 5. Adding Titles and Labels to Subplots
You can add titles and labels to each subplot individually, as well as a common title or label for the entire figure.


Annotations can be added to subplots to highlight specific data points, add labels, or create custom notes within your plots.


### a. Individual Titles and Labels
Each subplot can have its own title and axis labels.


```python

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plotting data with titles and labels
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('Y')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tangent Wave')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Y')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exponential')
axes[1, 1].set_xlabel('X')
axes[1, 1].set_ylabel('Y')

plt.tight_layout()

# Display the plot
plt.show()

```


### b. Adding a Common Title
You can add a common title for all subplots using `fig.suptitle()`.


```python

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plotting data with individual titles
axes[0, 0].plot(x, y1)
axes[0, 1].plot(x, y2)
axes[1, 0].plot(x, y3)
axes[1, 1].plot(x, y4)

# Adding a common title
fig.suptitle('Common Title for All Subplots', fontsize=16)

plt.tight_layout(rect=[0, 0, 1, 0.96])

# Display the plot
plt.show()

```


In this example:
- `fig.suptitle()` adds a title for the entire figure.
- `tight_layout(rect=[0, 0, 1, 0.96])` ensures that the subplots don’t overlap with the common title.



## 6. Inserting Inset Plots
An inset plot is a small plot within a larger plot, useful for zooming in on a particular area of interest.


```python

# Create main plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y1)
ax.set_title('Main Plot with Inset')

# Create an inset plot
inset_ax = fig.add_axes([0.5, 0.5, 0.4, 0.4])  # [left, bottom, width, height]
inset_ax.plot(x, y2)
inset_ax.set_title('Inset Plot')

# Display the plot
plt.show()

```


In this example:
- `fig.add_axes([0.5, 0.5, 0.4, 0.4])` adds an inset plot to the main figure.



## 7. Saving a Figure with Subplots
When saving a figure with subplots, it’s important to ensure that everything fits well within the saved image.


```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plotting data
axes[0, 0].plot(x, y1)
axes[0, 1].plot(x, y2)
axes[1, 0].plot(x, y3)
axes[1, 1].plot(x, y4)

# Adjust layout and add a common title
fig.suptitle('Figure with Subplots', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Save the figure
plt.savefig('figure_with_subplots.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```


In this example:
- `bbox_inches='tight'` ensures that all elements, including titles and labels, are included in the saved image.
