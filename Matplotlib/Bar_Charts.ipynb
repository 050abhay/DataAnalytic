### `matplotlib.pyplot.bar` 

The `matplotlib.pyplot.bar` function is used to create bar charts in Matplotlib, which are ideal for visualizing categorical data. A bar chart represents data with rectangular bars where the length of each bar is proportional to the value it represents. Bar charts can be either vertical or horizontal, and Matplotlib offers a lot of flexibility in customizing these plots.

## 1. Basic Usage of `bar` Function

### a. Creating a Simple Vertical Bar Chart

````python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

# Create a vertical bar chart
plt.bar(categories, values)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')

# Display the plot
plt.show()
````

## 2. Customizing Bars

### a. Changing Bar Colors

````python
plt.bar(categories, values, color='blue')
plt.title('Bar Chart with Custom Colors')
plt.show()
````

````python
colors = ['red', 'green', 'blue', 'orange']
plt.bar(categories, values, color=colors)
plt.title('Bar Chart with Different Colors')
plt.show()
````

### b. Adding Edge Colors

````python
plt.bar(categories, values, color='lightblue', edgecolor='black')
plt.title('Bar Chart with Edge Colors')
plt.show()
````

### c. Bar Width

````python
plt.bar(categories, values, width=0.5, color='purple')
plt.title('Bar Chart with Custom Bar Width')
plt.show()
````

## 3. Adding Labels and Titles

### a. Adding X and Y Labels

````python
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
````

## 4. Displaying Values on Bars

````python
bars = plt.bar(categories, values, color='skyblue')

# Adding value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')  # va: vertical alignment

plt.title('Bar Chart with Value Labels')
plt.show()
````

## 5. Horizontal Bar Chart

````python
plt.barh(categories, values, color='lightgreen')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.show()
````

## 6. Grouped and Stacked Bar Charts

### a. Grouped Bar Chart

````python
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 2, 5]
values2 = [2, 5, 8, 3]

# Positioning the bars
bar_width = 0.35
index = np.arange(len(categories))

plt.bar(index, values1, bar_width, label='Series 1', color='blue')
plt.bar(index + bar_width, values2, bar_width, label='Series 2', color='orange')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
plt.xticks(index + bar_width / 2, categories)  # Adjust x-ticks
plt.legend()

plt.show()
````

### b. Stacked Bar Chart

````python
plt.bar(categories, values1, color='blue', label='Series 1')
plt.bar(categories, values2, color='orange', bottom=values1, label='Series 2')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()

plt.show()
````

## 7. Customizing Ticks and Grids

### a. Customizing Ticks

````python
plt.bar(categories, values, color='coral')
plt.xticks(rotation=45)  # Rotate x-ticks
plt.title('Bar Chart with Rotated Ticks')
plt.show()
````

### b. Adding Grids

````python
plt.bar(categories, values, color='teal')
plt.grid(axis='y')  # Add gridlines to the y-axis
plt.title('Bar Chart with Gridlines')
plt.show()
````

## 8. Handling Missing Data

````python
import numpy as np

# Data with a missing value
values_with_nan = [3, np.nan, 2, 5]

plt.bar(categories, values_with_nan, color='lightcoral')
plt.title('Bar Chart with Missing Data')
plt.show()
````

## 9. Error Bars

````python
# Sample data with error bars
errors = [0.5, 0.4, 0.6, 0.3]

plt.bar(categories, values, yerr=errors, capsize=5, color='lightblue')
plt.title('Bar Chart with Error Bars')
plt.show()
````

## 10. Saving the Bar Chart

````python
plt.bar(categories, values, color='purple')
plt.title('Bar Chart to Save')
plt.savefig('bar_chart.png', dpi=300, bbox_inches='tight')
plt.show()
````

- Display value labels on bars, create grouped and stacked bar charts, and add error bars.

# Understanding the Logic

When you use the `groupby` function in pandas, the result is a `GroupBy` object. This object represents the grouped data, but no aggregation (like sum, mean, etc.) has been performed yet. The `GroupBy` object essentially holds the information on how the data is grouped but doesn't yet perform any calculations.

## Step-by-Step Breakdown

1. **`df.groupby("Payment Mode")`:**
    - This groups the DataFrame `df` by the "Payment Mode" column.
    - The result is a `GroupBy` object, which can be thought of as a collection of groups, where each group corresponds to a unique value in the "Payment Mode" column.
  
2. **`["Amount"]` after `groupby`:**
    - When you access `["Amount"]` after the `groupby` operation, you're selecting the "Amount" column within each group.
    - This doesn't alter the grouping; it just specifies that the subsequent operation (e.g., `sum()`, `mean()`) should be applied only to the "Amount" column within each group.
  
3. **`.sum()`:**
    - Finally, the `sum()` function is applied to each group, but only to the "Amount" column, as it was specified right before `.sum()`.
    - If you had selected multiple columns or didn't specify a column, you could aggregate across multiple columns simultaneously.

## Example to Clarify

Let's say you have the following DataFrame:

```python
import pandas as pd
data = {
        "Payment Mode": ["Credit Card", "Debit Card", "Cash", "Credit Card", "Cash"],
        "Amount": [120.5, 340.0, 260.3, 150.0, 90.7],
        "Customer": ["John", "Jane", "Jake", "John", "Jake"]
    }

df = pd.DataFrame(data)
```

Now, consider the two approaches:

1. Selecting a column after groupby:
    
```python
        grouped_by_amount = df.groupby("Payment Mode")["Amount"].sum()
        print(grouped_by_amount)
```
- This groups by `"Payment Mode"`, selects the `"Amount"` column, and then sums the amounts for each group.

2. Without specifying a column before aggregation:

```python
    grouped = df.groupby("Payment Mode").sum()
    print(grouped)
```
- This groups by `"Payment Mode"` and sums all numeric columns in the DataFrame (in this case, both "Amount" and potentially other numeric columns if they existed).

### Key Point:

- When you place `["Amount"]` after groupby, you're telling pandas to only perform the aggregation (like sum()) on the "Amount" column.
- If you don't specify a column like `["Amount"]`, pandas will try to aggregate all columns that it can (numeric ones by default).



````python
grouped_by= df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)

# Ploting these data 
plt.bar(grouped_by.index, grouped_by.values)

# # Adding labels and title
plt.xlabel('Payment Mode')
plt.ylabel('Amount in Rupee')
plt.title('Simple Bar Chart')

plt.show()
````
