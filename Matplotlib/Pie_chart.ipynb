
# matplotlib.pyplot.pie

The `matplotlib.pyplot.pie` function is used to create pie charts in Matplotlib. Pie charts are circular charts divided into sectors, each representing a proportion of the whole. They are particularly useful for visualizing categorical data and showing the relative proportions of different categories.

## 1. Basic Usage of pie Function
The `pie` function requires a sequence of numerical data (which represents the sizes of the pie pieces) and can optionally take other parameters like labels, colors, and explode.

### a. Simple Pie Chart


```python

import matplotlib.pyplot as plt

# Data
sizes = [25, 35, 20, 20]
labels = ['Category A', 'Category B', 'Category C', 'Category D']

# Create a simple pie chart
plt.pie(sizes, labels=labels)

# Adding a title
plt.title('Simple Pie Chart')

# Display the plot
plt.show()
```


In this example:

- `sizes` represents the size of each pie piece.
- `labels` represents the categories for each slice.

The `pie` function automatically calculates the proportions of each slice based on the sizes provided.

## 2. Customizing the Pie Chart
Matplotlib provides several options to customize the appearance of pie charts, including colors, explode effects, start angles, and more.

### a. Customizing Colors
You can customize the colors of the pie slices by passing a `colors` list.


```python

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors)
plt.title('Pie Chart with Custom Colors')
plt.show()
```


### b. Adding Explode Effect
The `explode` parameter allows you to "explode" or offset a slice from the pie chart for emphasis. This can be useful to highlight a specific category.


```python

explode = (0, 0.1, 0, 0)  # Explode the second slice (i.e., 'Category B')

plt.pie(sizes, labels=labels, colors=colors, explode=explode)
plt.title('Pie Chart with Explode Effect')
plt.show()
```


In this example, the second slice (Category B) is exploded outwards to draw attention to it.

### c. Starting Angle
The `startangle` parameter rotates the start of the pie chart by the given angle, which can help improve readability by positioning slices in a more intuitive way.


```python

plt.pie(sizes, labels=labels, colors=colors, startangle=90)
plt.title('Pie Chart with Start Angle')
plt.show()
```


## 3. Displaying Percentages
You can display the percentage of the total for each slice by using the `autopct` parameter.


```python

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Percentages')
plt.show()
```


In this example, the `autopct` parameter is set to `'%1.1f%%'`, which formats the percentages to one decimal place.

## 4. Adding Shadows
The `shadow` parameter adds a shadow effect to the pie chart, giving it a 3D appearance.


```python

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Pie Chart with Shadow')
plt.show()
```


## 5. Rotating the Pie Chart
In addition to setting the start angle, you can control the rotation direction of the pie chart using the `counterclock` parameter.


```python

plt.pie(sizes, labels=labels, colors=colors, startangle=90, counterclock=False)
plt.title('Pie Chart Rotated Clockwise')
plt.show()
```


In this example, setting `counterclock=False` rotates the pie chart clockwise instead of the default counterclockwise direction.

## 6. Customizing the Label Distance
The `labeldistance` parameter controls the radial distance at which the labels are drawn. You can adjust this to control how far the labels appear from the center of the pie chart.


```python

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, labeldistance=1.2)
plt.title('Pie Chart with Custom Label Distance')
plt.show()
```


In this example, the labels are drawn slightly further out from the center by setting `labeldistance=1.2`.

## 7. Customizing the Percentage Placement
The `pctdistance` parameter controls the position of the percentage labels relative to the center of the pie chart.


```python

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', pctdistance=0.85)
plt.title('Pie Chart with Custom Percentage Placement')
plt.show()
```


In this example, the percentage labels are drawn closer to the center by setting `pctdistance=0.85`.

## 8. Donut (Ring) Chart
You can create a donut chart, which is a variation of a pie chart with a hole in the center, by using the `wedgeprops` parameter to set the width of the slices.


```python

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.3})
plt.title('Donut Chart')
plt.show()
```


In this example, the `wedgeprops={'width': 0.3}` creates a ring with an inner radius of 70% of the pie's radius.

## 9. Exploding All Slices
If you want to explode all slices of the pie chart, you can specify a larger list for the `explode` parameter.


```python

explode = [0.1] * len(sizes)  # Explode all slices equally

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode)
plt.title('Pie Chart with All Slices Exploded')
plt.show()
```


## 10. Adding Legends
Sometimes, the pie chart labels can become cluttered, especially when using small slices or displaying percentages. In such cases, you can remove the labels from the chart itself and use a legend instead.


```python

plt.pie(sizes, colors=colors, startangle=140)
plt.legend(labels, loc="best")
plt.title('Pie Chart with Legend')
plt.show()
```


## 11. Handling Missing or Zero Values
Pie charts can handle zero values gracefully, and such slices won't appear in the chart.


```python

sizes_with_zero = [25, 0, 20, 20]  # One category with a zero value

plt.pie(sizes_with_zero, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Zero Value')
plt.show()
```


In this example, the second category is given a zero value, so it will not appear in the chart.

## 12. Saving the Pie Chart
You can save the pie chart to a file using `plt.savefig()`.


```python

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Pie Chart to Save')
plt.savefig('pie_chart.png', dpi=300, bbox_inches='tight')
plt.show()
```


## 13. Pie Chart Subplots
You can create multiple pie charts in the same figure using `plt.subplot()`.


```python

plt.figure(figsize=(10, 5))

# First pie chart
plt.subplot(1, 2, 1)
plt.pie(sizes, labels=labels, colors=colors, startangle=90)
plt.title('Pie Chart 1')

# Second pie chart
plt.subplot(1, 2, 2)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Pie Chart 2')

plt.tight_layout()
plt.show()
```


## 14. Advanced Customization with `wedgeprops` and `textprops`
The `wedgeprops` parameter allows you to customize the appearance of the pie slices, such as their edge color and width, while `textprops` allows you to customize the text properties.


```python

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True,
        startangle=140, wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
        textprops={'fontsize': 14, 'color': 'navy'})
plt.title('Pie Chart with Custom Wedge and Text Properties')
plt.show()
```


In this example:

- `wedgeprops` adds a black edge with a linewidth of 1.5 to the slices.
- `textprops` sets the font size to 14 and the text color to navy for the labels.

## Summary:
The `matplotlib.pyplot.pie` function is a powerful tool for creating pie charts in Matplotlib. It offers:

- **Basic Plotting**: Simple pie charts to visualize proportions of categorical data.
- **Customization**: Extensive control over colors, explode effects, start angles, labels, and shadows.
- **Advanced Features**: Ability to display percentages, handle missing or zero values, create donut charts, and add legends.
- **Subplots**: Create multiple pie charts in the same figure.
- **Saving Plots**: Easily save the pie charts to files for later use.

Pie charts are ideal for presenting a visual summary of data in a way that is easy to understand, especially when dealing with proportional data. Matplotlib’s `pie` function provides all the tools needed to create informative and visually appealing pie charts.
