# Plotting in Matplotlib Overview

Plotting in Matplotlib is a broad topic due to its versatility and depth. Below, I’ll provide a detailed overview of plotting in Matplotlib, covering the basics, customization options, types of plots, and more advanced techniques.

## 1. Basic Plotting with pyplot
Matplotlib’s pyplot module provides a MATLAB-like interface for creating simple plots quickly. It’s the most commonly used interface in Matplotlib and is ideal for generating basic plots.

### a. Basic Line Plot
A line plot is one of the most basic types of plots. It is used to visualize the relationship between two variables.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Creating a line plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Basic Line Plot')

# Display the plot
plt.show()
```

### b. Plotting Multiple Lines
You can plot multiple lines on the same graph by calling `plt.plot()` multiple times.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 35]
y2 = [15, 25, 30, 35, 40]

# Plotting multiple lines
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')

# Adding labels, title, and legend
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Multiple Line Plot')
plt.legend()

# Display the plot
plt.show()
```

You can customize the appearance of lines using parameters like color, linestyle, and linewidth.

```python
plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o')
```

### b. Adding Labels and Titles
Labels and titles are essential for making your plots understandable.

```python


plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Plot Title')
```

### d. Adding Legends
Legends are useful for identifying different lines or plots.

```python
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend(loc='upper left')
```

### e. Annotations
You can add annotations to highlight specific points on the plot.

```python
plt.annotate('Max Value', xy=(5, 35), xytext=(3, 35),
             arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                # Sine function

# Create a plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Sine Wave')

# Add a specific annotation
highlight_x = np.pi  # Point where x = π
highlight_y = np.sin(highlight_x)  # y value at x = π

# Plot the highlighted point
plt.plot(highlight_x, highlight_y, 'ro')  # 'ro' means red color and circle marker

# Annotate the highlighted point
plt.annotate('Peak at π',               # Annotation text
             xy=(highlight_x, highlight_y),  # Coordinates to annotate
             xytext=(highlight_x + 1, highlight_y + 0.5),  # Coordinates for text
             arrowprops=dict(facecolor='black', arrowstyle='->'))  # Arrow properties

# Set title and labels
plt.title('Sine Wave with Annotation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.show()
```

## 3. Types of Plots
Matplotlib supports a wide variety of plots, each suited for different types of data and analyses.

### a. Bar Plots
Bar plots are used to represent categorical data with rectangular bars.

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

plt.bar(categories, values)
plt.title('Bar Plot')
plt.show()
```

```python
plt.barh(categories, values)
```

### b. Histograms
Histograms are used to represent the distribution of numerical data.

```python
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.5, color='blue')
plt.title('Histogram')
plt.show()
```

### c. Scatter Plots
Scatter plots are used to visualize the relationship between two continuous variables.

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green')
plt.title('Scatter Plot')
plt.show()
```

### d. Pie Charts
Pie charts represent the proportion of categorical data as slices of a pie.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

### e. Box Plots
Box plots are used to visualize the distribution of data based on a five-number summary.

```python
data = [np.random.randn(100) for _ in range(4)]
plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

### f. Heatmaps
Heatmaps are used to visualize matrix-like data, often used for showing correlations or distributions.

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```

## 4. Advanced Plotting Techniques

### a. Subplots
You can create multiple plots in the same figure using `plt.subplot()`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 50)  # 50 points between 0 and 10
y1 = np.sin(x)  # y1 has 50 points
y2 = np.cos(x)  # y2 has 50 points

plt.figure(figsize=(10, 5))

# First subplot
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title('Subplot 1')

# Second subplot
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Subplot 2')

plt.tight_layout()
plt.show()
```

* **Explanation:**
- `x = np.linspace(0, 10, 50)` creates an array x with 50 points between 0 and 10.
- `y1 = np.sin(x) and y2 = np.cos(x)` generate y1 and y2 arrays with the same length as x, which avoids the dimension mismatch error.
- The `plt.subplot(1, 2, 1)` and `plt.subplot(1, 2, 2)` create two subplots in a 1x2 grid.
- Finally, `plt.tight_layout()` adjusts the spacing between subplots to prevent overlap.

```python
# This code will create 5 subplots arranged in a 2x3 grid. The sixth position in the grid will be empty. 
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x / 10)
y5 = np.log(x + 1)

plt.figure(figsize=(15, 10))

# First subplot
plt.subplot(2, 3, 1)
plt.plot(x, y1)
plt.title('Subplot 1: Sine')

# Second subplot
plt.subplot(2, 3, 2)
plt.plot(x, y2)
plt.title('Subplot 2: Cosine')

# Third subplot
plt.subplot(2, 3, 3)
plt.plot(x, y3)
plt.title('Subplot 3: Tangent')

# Fourth subplot
plt.subplot(2, 3, 4)
plt.plot(x, y4)
plt.title('Subplot 4: Exponential')

# Fifth subplot
plt.subplot(2, 3, 5)
plt.plot(x, y5)
plt.title('Subplot 5: Logarithm')

plt.tight_layout()
plt.show()
```

### d. Working with Dates
Matplotlib can handle date-time data using the `dates` module.

```python
import matplotlib.dates as mdates
import datetime

dates = [datetime.date(2023, 1, i) for i in range(1, 10)]
values = np.random.rand(9)

plt.plot(dates, values)
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.title('Date Plot')
plt.show()
```

## 7. Interactive Plotting
Matplotlib supports interactive plotting, especially in environments like Jupyter Notebooks. You can enable interactive mode using:

```python
plt.ion()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

# Enable interactive mode
plt.ion()

# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create an initial plot
fig, ax = plt.subplots()
line, = ax.plot(x, y)

# Update the plot interactively
for phase in np.linspace(0, 2 * np.pi, 100):
    line.set_ydata(np.sin(x + phase))  # Update the y-data of the plot
    fig.canvas.draw()  # Redraw the figure
    fig.canvas.flush_events()  # Process events to update the plot
    time.sleep(0.1)  # Pause to see the updates

# Keep the plot open
# plt.ioff()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

# Turn on interactive mode
plt.ion()

# Create a sine wave plot
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
line, = plt.plot(x, y)

# Update the plot in a loop
for shift in np.linspace(0, 2 * np.pi, 100):
    y = np.sin(x + shift)  # Shift the sine wave
    line.set_ydata(y)  # Update the plot with the new data
    plt.draw()  # Redraw the plot
    time.sleep(0.1)  # Pause briefly to see the update

# # Turn off interactive mode and keep the final plot open
plt.ioff()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

# Turn on interactive mode
plt.ion()

# Create a sine wave plot
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
line, = plt.plot(x, y)
plt.ylim(-1.5, 1.5)  # Set limits for y-axis

# Update the plot in a loop
for shift in np.linspace(0, 2 * np.pi, 100):
    y = np.sin(x + shift)  # Shift the sine wave
    line.set_ydata(y)  # Update the plot with the new data
    plt.draw()  # Redraw the plot
    plt.pause(0.1)  # Pause briefly to see the update

# Turn off interactive mode and keep the final plot open
plt.ioff()
plt.show()
```

## 8. Object-Oriented Interface
While `pyplot` is easy for quick plots, Matplotlib also provides an object-oriented interface for more complex and fine-grained control.

### a. Creating Figures and Axes

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Object-Oriented Plot')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
plt.show()
```

### b. Multiple Axes and Layouts
You can create complex layouts with multiple subplots using the object-oriented approach.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y1 = np.sin(x)                # y1 has 100 points
y2 = np.cos(x)                # y2 also has 100 points

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot on the first subplot
ax1.plot(x, y1)
ax1.set_title('Sine Wave')

# Plot on the second subplot
ax2.plot(x, y2)
ax2.set_title('Cosine Wave')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
