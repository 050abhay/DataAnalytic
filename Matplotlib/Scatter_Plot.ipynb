
3. **matplotlib.pyplot.scatter**

The `matplotlib.pyplot.scatter` function is used to create scatter plots in Matplotlib. Scatter plots are a type of plot that shows individual data points on a two-dimensional graph, where the position of each point is determined by two variables. Scatter plots are particularly useful for visualizing the relationship between two continuous variables, identifying trends, clusters, and outliers.



1. **Basic Usage of scatter Function**

The scatter function requires two main inputs: the x-values and the y-values. Each pair of x and y values corresponds to a single point on the scatter plot.


`````python

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a simple scatter plot
plt.scatter(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')

# Display the plot
plt.show()
`````


In this example:
- `x` represents the data for the x-axis.
- `y` represents the data for the y-axis.

The scatter function plots each pair of x and y values as a point on the graph.



2. **Customizing Markers**

You can customize the appearance of the points (markers) on the scatter plot using parameters such as color, marker, `s` (size), and `alpha` (transparency).



a. **Changing Marker Color**

You can change the color of the markers using the `color` parameter.


`````python

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Changing Marker Color
plt.scatter(x, y, color='red')
plt.title('Scatter Plot with Red Markers')
plt.show()
`````


b. **Changing Marker Size**

The `s` parameter controls the size of the markers. You can pass a single value to set a uniform size for all markers, or an array to vary the size based on a third variable.


`````python

# Changing Marker Size
plt.scatter(x, y, s=100)  # Larger markers
plt.title('Scatter Plot with Larger Markers')
plt.show()
`````

`````python

sizes = [20, 50, 100, 200, 300]  # Variable marker sizes
plt.scatter(x, y, s=sizes)
plt.title('Scatter Plot with Variable Marker Sizes')
plt.show()
`````


c. **Changing Marker Shape**

The `marker` parameter allows you to change the shape of the markers. Some common options include:
- `'o'` (circle, default)
- `'s'` (square)
- `'^'` (triangle up)
- `'D'` (diamond)
- `'x'` (cross)


`````python

# Changing Marker Shape
plt.scatter(x, y, marker='^')  # Triangle markers
plt.title('Scatter Plot with Triangle Markers')
plt.show()
`````


d. **Adjusting Transparency**

The `alpha` parameter controls the transparency of the markers. A value of 0 is fully transparent, while 1 is fully opaque.


`````python

# Adjusting Transparency
plt.scatter(x, y, color='blue', alpha=0.5)  # Semi-transparent markers
plt.title('Scatter Plot with Transparency')
plt.show()
`````


3. **Using Different Colors for Points**

You can assign different colors to the points based on a third variable using the `c` parameter.


`````python

import matplotlib.pyplot as plt
import numpy as np

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
colors = [10, 20, 30, 40, 50]  # Third variable determining colors

plt.scatter(x, y, c=colors, cmap='viridis')  # Use a colormap
plt.colorbar()  # Show color scale
plt.title('Scatter Plot with Color Mapping')
plt.show()
`````


In this example:
- `c` parameter is used to specify the colors of the points.
- `cmap` parameter specifies the colormap to use (e.g., 'viridis', 'plasma', 'inferno', 'magma', etc.).
- `plt.colorbar()` adds a color bar to indicate the color mapping.



4. **Advanced Scatter Plot Customizations**



a. **Combining Marker Size and Color**

You can vary both the size and color of the markers to represent additional dimensions of data.


`````python

sizes = [20, 50, 100, 200, 300]
colors = [10, 20, 30, 40, 50]

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)
plt.colorbar()
plt.title('Scatter Plot with Variable Size and Color')
plt.show()
`````


b. **Adding Labels to Points**

You can annotate individual points on the scatter plot with text labels using `plt.text()`.


`````python

plt.scatter(x, y, color='green')

# Adding labels to points
for i, label in enumerate(['A', 'B', 'C', 'D', 'E']):
    plt.text(x[i], y[i], label)

plt.title('Scatter Plot with Labels')
plt.show()
`````


5. **Scatter Plot with Multiple Data Series**

You can plot multiple datasets on the same scatter plot by calling `plt.scatter()` multiple times.


`````python

# Data for two groups
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]

x2 = [1, 2, 3, 4, 5]
y2 = [3, 6, 9, 12, 15]

# Scatter plot for both groups
plt.scatter(x1, y1, color='blue', label='Group 1')
plt.scatter(x2, y2, color='orange', label='Group 2')

# Adding labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Multiple Data Series')
plt.legend()

plt.show()
`````


6. **Logarithmic and Other Scales**

You can change the scale of the axes to logarithmic, symlog, or other types using `plt.xscale()` and `plt.yscale()`.


`````python

plt.scatter(x, y, color='purple')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Logarithmic X-axis')
plt.ylabel('Logarithmic Y-axis')
plt.title('Scatter Plot with Logarithmic Scales')
plt.show()
`````


7. **Handling Large Datasets**

For large datasets, scatter plots can become crowded. Here are some techniques to handle large datasets:



a. **Reducing Marker Size and Transparency**

Reducing the marker size and increasing transparency can help visualize dense plots.


`````python

import matplotlib.pyplot as plt
import numpy as np

# Data for large dataset
large_x = np.random.rand(1000)
large_y = np.random.rand(1000)

# Reducing Marker Size and Transparency
plt.scatter(large_x, large_y, s=10, alpha=0.3)
plt.title('Scatter Plot for Large Datasets')
plt.show()
`````


b. **Hexbin Plot**

For very large datasets, consider using a hexbin plot, which aggregates data into hexagonal bins.


`````python

# Hexbin Plot
plt.hexbin(large_x, large_y, gridsize=30, cmap='Blues')
plt.colorbar(label='Count')
plt.title('Hexbin Plot for Large Datasets')
plt.show()
`````


8. **Subplots with Scatter Plots**

You can create multiple scatter plots in a single figure using `plt.subplot()`.


`````python

# Subplots with Scatter Plots
plt.figure(figsize=(10, 5))

# First subplot
plt.subplot(1, 2, 1)
plt.scatter(large_x, large_y, color='blue')
plt.title('Scatter Plot 1')

# Second subplot
plt.subplot(1, 2, 2)
plt.scatter(large_y, large_x, color='red')
plt.title('Scatter Plot 2')

plt.tight_layout()
plt.show()
`````


9. **Interactive Scatter Plots**

In interactive environments like Jupyter Notebooks, you can create scatter plots that allow for dynamic exploration.


`````python
%matplotlib inline
plt.scatter(x, y, color='green')
plt.title('Interactive Scatter Plot')
plt.show()
`````


10. **Saving the Scatter Plot**

You can save the scatter plot to a file using `plt.savefig()`.


`````python

# Saving the Scatter Plot
plt.scatter(large_x, large_y, color='purple')
plt.title('Scatter Plot to Save')
plt.savefig('scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show()
`````


11. **3D Scatter Plots**

Matplotlib supports 3D scatter plots through the `mplot3d` toolkit.


`````python

from mpl_toolkits.mplot3d import Axes3D

# Data for 3D scatter plot
z = [1, 4, 9, 16, 25]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(large_x[:5], large_y[:5], z, color='teal')

# Adding labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')

plt.show()
`````

The line of code:
````python
plt.scatter(df["Annual Salary"], df["EEID"], s=df["Age"] * 10, c='blue', alpha=0.5)
````
creates a scatter plot using Matplotlib, where each point on the plot represents data from the DataFrame df. Heres a detailed explanation of each component:

1. **plt.scatter(x, y, s, c, alpha)**

The plt.scatter() function creates a scatter plot. It plots points in a two-dimensional space, where:
- `x` defines the positions of the points along the horizontal axis.
- `y` defines the positions of the points along the vertical axis.
- `s` controls the size of the points.
- `c` sets the color of the points.
- `a`lpha sets the transparency of the points.

2. **df["Annual Salary"]**
- Purpose: This selects the data from the "Annual Salary" column in the DataFrame df.
- Role in Plot: These values are used for the x-coordinates of the scatter plot. Each point's horizontal position is determined by the corresponding value in this column.

3. **df["EEID"]**
- Purpose: This selects the data from the "EEID" (Employee ID) column in the DataFrame df.
- Role in Plot: These values are used for the y-coordinates of the scatter plot. Each point's vertical position is determined by the corresponding value in this column.

4. **s=df["Age"] * 10**
- Purpose: This controls the size of each point on the scatter plot.
- Explanation:
    - df["Age"] selects the data from the "Age" column in the DataFrame df.
    - The * 10 part scales the values by a factor of 10. This is done because the default point size might be too small if you used the raw "Age" values. Multiplying by 10 makes the    points larger and more visible.

- Role in Plot: The size of each point in the scatter plot is proportional to the employee's age. An older employee (higher age value) will have a larger point on the plot.

5. **c='blue'**
- Purpose: This sets the color of all the points on the scatter plot.
- Explanation:
    - c='blue' tells Matplotlib to color all points blue.
- Role in Plot: Every point on the scatter plot will appear in blue. You can also use other color names (e.g., 'red', 'green'), hexadecimal color codes (e.g., '#1f77b4'), or even use a list or array to specify different colors for different points.

6. **alpha=0.5**
- Purpose: This controls the transparency (or opacity) of the points on the scatter plot.
- Explanation:
    - alpha can take any value between 0 and 1.
    - alpha=1 means the points are fully opaque (no transparency).
    - alpha=0 means the points are fully transparent (invisible).
    - alpha=0.5 means the points are semi-transparent.
- Role in Plot: The transparency is useful if points overlap. With alpha=0.5, overlapping points will appear darker where they overlap, which can help in visualizing dense regions in the data.
  
**Final Output**

The final output of this line of code is a scatter plot where:
- The x-axis represents employee annual salaries.
- The y-axis represents employee IDs.
- The size of each point corresponds to the employee's age, scaled by a factor of 10.
- All points are colored blue.
- Points are semi-transparent, allowing for better visibility when points overlap.
