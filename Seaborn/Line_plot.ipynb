# seaborn.lineplot
The `seaborn.lineplot` function is a powerful tool for creating line plots, which are used to display data points in a continuous sequence, typically representing trends over time or the relationship between two variables. Seabornâ€™s lineplot builds on top of Matplotlib, offering a high-level interface that simplifies the creation of aesthetically pleasing and informative line plots.

## 1. Basic Usage of seaborn.lineplot
The lineplot function is typically used to plot data where one variable is continuous and another can be either continuous or categorical.

### Example: Simple Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Time series
data = sns.load_dataset('flights')

# Create a line plot
sns.lineplot(x='year', y='passengers', data=data)

# Display the plot
plt.show()
```

In this example:
- `x='year'`: The year is plotted on the x-axis.
- `y='passengers'`: The number of passengers is plotted on the y-axis.
- `data=data`: The flights dataset is used.

## 2. Plotting Multiple Lines
You can plot multiple lines on the same plot by using the `hue`, `style`, and `size` parameters to differentiate between lines.

### Example: Multiple Lines with hue

```python
sns.lineplot(x='year', y='passengers', hue='month', data=data)
plt.show()
```

- `hue='month'`: This adds color coding to differentiate between the months, with each line representing a different month.

### Example: Multiple Lines with style

```python
sns.lineplot(x='year', y='passengers', hue='month', style='month', data=data)
plt.show()
```

- `style='month'`: This parameter changes the line style for each month, such as dashed or dotted lines.

## 3. Using ci for Confidence Intervals
The `ci` parameter controls the size of the confidence interval around the line.

- **Default**: A 95% confidence interval is plotted by default.
- `ci=None`: No confidence interval is plotted.
- `ci='sd'`: The standard deviation of the data is shown as the confidence interval.

### Example: Plot with Confidence Interval

```python
sns.lineplot(x='year', y='passengers', ci='sd', data=data)
plt.show()
```

The warning we received indicates that the `ci` parameter in sns.lineplot is deprecated and will be removed in future versions of Seaborn. Instead, we should use the `errorbar` parameter to achieve the same effect.

Here's the updated code:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Time series
data = sns.load_dataset('flights')

# Plotting with the updated parameter
sns.lineplot(x='year', y='passengers', errorbar='sd', data=data)
plt.show()
```

- `errorbar='sd'` replaces the `ci='sd'` parameter. It instructs Seaborn to display the standard deviation as error bars on the line plot.

## 4. Adding Markers
You can add markers to the data points using the `markers=True` parameter, which makes it easier to see the individual data points on the line.

### Example: Line Plot with Markers

```python
sns.lineplot(x='year', y='passengers', markers=True, data=data)
plt.show()
```

## 5. Handling Missing Data
Seabornâ€™s lineplot can handle missing data points gracefully by using the `nan_policy` parameter.

- `nan_policy='drop'`: Drops missing values (default behavior).
- `nan_policy='propagate'`: Propagates missing values, leading to gaps in the line.

## 6. Customizing Line Appearance
You can customize the appearance of the lines, such as color, line width, and transparency, using various parameters:

- `color='color_name'`: Sets the color of the line.
- `linewidth=float`: Sets the width of the line.
- `alpha=float`: Controls the transparency of the line.

### Example: Customized Line Appearance

```python
sns.lineplot(x='year', y='passengers', data=data, color='red', linewidth=2.5, alpha=0.8)
plt.show()
```

## 7. Facet Grids with lineplot
FacetGrid allows you to create multiple line plots by splitting the data across multiple facets.

### Example: Faceted Line Plot

```python
g = sns.FacetGrid(data, col='month', col_wrap=4, height=3)
g.map(sns.lineplot, 'year', 'passengers')
plt.show()
```

- `col='month'`: Creates a separate plot for each month.
- `col_wrap=4`: Limits the number of columns to 4 before wrapping to the next row.
- `height=3`: Sets the height of each facet.

## 8. Handling Time Series Data
When plotting time series data, itâ€™s often useful to convert your time data into a datetime object using Pandas.

### Example: Time Series Line Plot

```python
import pandas as pd

# Convert 'year' to a datetime object
data['year'] = pd.to_datetime(data['year'], format='%Y')

sns.lineplot(x='year', y='passengers', data=data)
plt.show()
```

This conversion allows you to take advantage of Seabornâ€™s built-in time series handling capabilities.

## 9. Plotting with DataFrames
Seaborn works seamlessly with Pandas DataFrames. You can easily pass a DataFrame and specify the column names for the x and y axes.

### Example: Line Plot from a DataFrame

```python
data = pd.DataFrame({
    'x': range(10),
    'y': [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
})

sns.lineplot(x='x', y='y', data=data)
plt.show()
```

## 10. Combining lineplot with Other Seaborn Functions
You can combine `lineplot` with other Seaborn functions to create more complex visualizations, such as overlaid distributions, heatmaps, or scatterplots.

### Example: Overlaying a Distribution

```python
# Sample data: Time series
data = sns.load_dataset('flights')
#Plotting
sns.distplot(data['passengers'], hist=False, kde=True)
sns.lineplot(x='year', y='passengers', data=data)
plt.show()
```

- `sns.distplot` is deprecated and should be replaced with either `sns.kdeplot` or `sns.displot`.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Time series
data = sns.load_dataset('flights')

# Plotting
sns.kdeplot(data['passengers'], bw_adjust=0.5)  # Replacing distplot with kdeplot
sns.lineplot(x='year', y='passengers', data=data)
plt.show()
```

- `sns.kdeplot` is used for kernel density estimation (KDE) and replaces the deprecated `sns.distplot`. The `bw_adjust` parameter can be used to adjust the bandwidth of the KDE, but itâ€™s optional.
- **KeyError Fix**: The correct column names ('year', 'month', 'passengers') are used to avoid the `KeyError`.

## 11. Advanced Example: Multiple Customizations
Hereâ€™s an example that combines several features of `lineplot`:

```python
sns.lineplot(
x="year", 
y="passengers", 
hue="month", 
style="month", 
markers=True, 
dashes=False, 
data=data, 
palette="tab10", 
linewidth=2.5
)
plt.title("Number of Passengers Over Time")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend(title="Month")
plt.show()
                   
```


 ## Summary of Important Parameters

- **x and y**: Variables to be plotted on the x and y axes.
- **hue**: Variable that defines subsets of the data, with different colors.
- **style**: Variable that defines different line styles.
- **size**: Variable that controls the size of the lines.
- **markers**: Adds markers at each data point.
- **ci**: Confidence interval to plot (default is 95%).
- **linewidth**: Thickness of the lines.
- **alpha**: Transparency of the lines.
- **palette**: Color palette to use for different levels of the hue variable.
